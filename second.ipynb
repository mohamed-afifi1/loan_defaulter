```python
def drop_high_nulls(df, threshold=45):

    # Calculate percentage of missing values
    null_percent = (df.isnull().sum() / len(df)) * 100

    # Find columns to drop
    cols_to_drop = null_percent[null_percent >= threshold].index

    # Print info
    print(f" Columns with  {threshold}% missing values:")
    print(null_percent[null_percent >= threshold].sort_values(ascending=False))
    print(f"\n Dropping {len(cols_to_drop)} columns...")

    # Drop and return
    df_application = df.drop(columns=cols_to_drop)
    print(f" Cleaned DataFrame shape: {df_application.shape}")

    return df_application
```

```python
df_application = drop_high_nulls(df_application, threshold=45)
```

```python
pd.set_option('display.max_rows', None)

(df_application.isnull().mean() * 100).round(2).sort_values(ascending=False)
```

```python
# Calculate null percentages for each column
null_percent = (df_application.isnull().sum() / len(df_application)) * 100

# Select columns that have < 1% missing values
cols_with_low_nulls = null_percent[null_percent < 1].index

# Drop rows with NaN in those columns
df_application = df_application.dropna(subset=cols_with_low_nulls)
```

```python
def drop_high_nulls(df, threshold=40):

    # Calculate percentage of missing values
    null_percent = (df.isnull().sum() / len(df)) * 100

    # Find columns to drop
    cols_to_drop = null_percent[null_percent >= threshold].index

    # Print info
    print(f" Columns with  {threshold}% missing values:")
    print(null_percent[null_percent >= threshold].sort_values(ascending=False))
    print(f"\n Dropping {len(cols_to_drop)} columns...")

    # Drop and return
    df_pervious = df.drop(columns=cols_to_drop)
    print(f" Cleaned DataFrame shape: {df_pervious.shape}")

    return df_pervious
```
