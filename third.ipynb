```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns' , None)
```

```python
df1 = pd.read_csv('df_application(no null).csv')
df1
```

```python
df1.head()
```

```python
df2 = pd.read_csv('df_previous(no null).csv')
df2
```

```python
df2.head()
```

```python
df = pd.concat([df1, df2] , ignore_index=True)
df
```

```python
df.head()
```

```python
cols = ['AMT_INCOME_TOTAL' , 'AMT_CREDIT' , 'AMT_ANNUITY' , 'AMT_GOODS_PRICE']
for col in cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    LB = Q1 - 1.5 * IQR
    UP = Q3 + 1.5 * IQR
    less_than = (df[col] < LB).sum()
    more_than = (df[col] > UP).sum()
    print(f"less than lower bound {less_than}")
    print(f"more than upper bound {more_than}")
    print(f"Max = {df[col].max()} UP = {UP}")
    print(f"Min = {df[col].min()} LB = {LB}")
    print('---------------------------------------')
```
