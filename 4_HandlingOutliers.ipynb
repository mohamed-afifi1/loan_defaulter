```python
cols = ['AMT_INCOME_TOTAL' , 'AMT_CREDIT', 'AMT_ANNUITY' , 'AMT_GOODS_PRICE' , 'REGION_POPULATION_RELATIVE' , 'EXT_SOURCE_2' , 'EXT_SOURCE_3' , 'YEARS_BIRTH'
        , 'YEARS_EMPLOYED' , 'YEARS_REGISTRATION' , 'YEARS_ID_PUBLISH' , 'YEARS_LAST_PHONE_CHANGE' , 'AMT_ANNUITY_PREVIOUS' , 'AMT_APPLICATION' , 'AMT_CREDIT_PREVIOUS' , 'AMT_GOODS_PRICE_PREVIOUS' , 
       'YEARS_DECISION' , 'YEARS_FIRST_DRAWING' , 'YEARS_FIRST_DUE' , 'YEARS_LAST_DUE_1ST_VERSION' , 'YEARS_LAST_DUE' , 'YEARS_TERMINATION']
for col in cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    LB = Q1 - 1.5 * IQR
    UP = Q3 + 1.5 * IQR
    less_than = (df[col] < LB).sum()
    more_than = (df[col] > UP).sum()
    print(col)
    print(f"less than lower bound {less_than}")
    print(f"more than upper bound {more_than}")
    print(f"Max = {df[col].max()} UP = {UP}")
    print(f"Min = {df[col].min()} LB = {LB}")
    print('---------------------------------------')
```
